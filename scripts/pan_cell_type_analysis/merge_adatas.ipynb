```python
import scanpy as sc


def bringCellTypes_fromRegions(adatas_path, cell_types, column_of_cell_type, region_index=None):
    adatas = list()

    for adata_xenium_p in adatas_path:    
        
        adata_chosen = sc.read_h5ad(adata_xenium_p)
        mask = adata_chosen.obs[column_of_cell_type].isin(cell_types)

        adatas.append(adata_chosen[mask].copy())

        adatas_merged = sc.concat(
        adatas,
        join="inner",             # keep only shared genes
        label="Region",            # name of the obs column marking original dataset
        keys = [f"R{region_index[i]}" if region_index is not None else f"R{i}" for i in range(1, len(adatas_path) + 1)]
    )

    adatas_merged.obs[column_of_cell_type] = (adatas_merged.obs[column_of_cell_type].astype(str) + "_" + adatas_merged.obs["Region"].astype(str))

    return adatas_merged
    
```

```python
adatas_path = []
for Tissue,TableFolder in [["Region1","REGION1_TABLES"],["Region2","REGION2_TABLES"],["Region3","REGION3_TABLES"],["Region4", "REGION4_TABLES"]]:
    in_out_path = f"/Volumes/ProstateCancerEvoMain/dbs/Ongoing/{Tissue}/{TableFolder}"
    adatas_path.append(f"{in_out_path}/{Tissue}_Xenium_Phen_HE_Integrated.Xenium_Process_Table.Cell_Type_Annotated_Auto.V3.All_Genes.h5ad")


adata_concat = bringCellTypes_fromRegions(adatas_path, ["tumor_cell_markers","mature_B_cell"], "cell_type_xenium_panel")
```
